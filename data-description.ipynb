# Duration of audio

```python
import librosa
from pathlib import Path
import pandas as pd
import statistics
from statistics import mean 
```

```python
patients_dur = []
control_dur = []
```

```python
def get_duraction(file_path, ls):
    duraction = librosa.get_duration(path=file_path)
    ls.append(duraction)
```

```python
p = Path("/kaggle/input/patients")

for x in p.iterdir():
    get_duraction(str(x), patients_dur) 
    
p = Path("/kaggle/input/control")

for x in p.iterdir():
    get_duraction(str(x), control_dur) 
```

```python
print(f"""Control: max={max(control_dur)}, min={min(control_dur)},
      mean={mean(control_dur)}, SD={statistics.pstdev(control_dur)}""")
```

```python
print(f"""Patients: max={max(patients_dur)}, min={min(patients_dur)},
      mean={mean(patients_dur)}, SD={statistics.pstdev(patients_dur)}""")
```

# Duration of new audio

```python
patients_dur_new = []
control_dur_new = []

p = Path("/kaggle/input/patients-prerocessed")

for x in p.iterdir():
    get_duraction(str(x), patients_dur_new) 
    
p = Path("/kaggle/input/control-preprocessed")

for x in p.iterdir():
    get_duraction(str(x), control_dur_new) 
    
print(f"""Control: max={max(control_dur_new)}, min={min(control_dur_new)},
      mean={mean(control_dur_new)}, SD={statistics.pstdev(control_dur_new)}""")

print(f"""Patients: max={max(patients_dur_new)}, min={min(patients_dur_new)},
      mean={mean(patients_dur_new)}, SD={statistics.pstdev(patients_dur_new)}""")
```
